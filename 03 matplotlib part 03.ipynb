# Plot appearance

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,11)
y=x**2
```

```python
x
```

```python
y
```

```python
fig= plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='#FF0000')  # RGB hex colour 
axes.plot(x,y,color='blue',linewidth=10,alpha=0.7)  # RGB hex colour   , alpha for transparant , above line code is displaying 
axes.plot(y,x)
```

```python
fig= plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',lw=3, linestyle='-.')  # lw = linewidth
axes.plot(x,x**2,color='red',lw=3, linestyle=':')
axes.plot(x,x**3,color='red',lw=3, linestyle='steps')
axes.plot(x,x**4,color='red',lw=3, ls='-')    #ls=linestyle
```

```python
fig= plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',lw=1, ls='-',marker='o', markersize=20,markerfacecolor='green')   # + ,p ,o,*,1
axes.plot(x,y+10,color='red',lw=1, ls='-',marker='o', markersize=20,markerfacecolor='yellow',markeredgewidth=5,markeredgecolor='green')   # markeredgewidth

```

# X limit and Y LIMIT

```python
fig= plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=2, ls='-')

axes.set_xlim(0,15)
axes.set_5lim(0,150)
```
