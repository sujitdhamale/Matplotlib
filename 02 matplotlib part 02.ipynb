# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,11)
y=x**2
```

```python
x
```

```python
y
```

## subplot using Object oriented method 

```python
fig,axes=plt.subplots(nrows=2, ncols=2)
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
```

```python
axes   # Axis is array and it is iterable object
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
for curr_axis in axes:
    curr_axis.plot(x,y)
```


--------------------------







### Using Index also we can access AXIS

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

# Figure Size , Aspect ratio and DPI

*  figsize : tuple of integers,width, height in inches.
*  dpi : (dots per inches/Pixel per inches )integer,resolution of the figure. defaults to rc figure.dpi.


```python
fig=plt.figure(figsize=(8,2), dpi=None)

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[0].set_title("Title1")

axes[1].plot(y,x)
axes[1].set_title("Title2")
plt.tight_layout()
```

# legend

###  The legend is linked to the data being graphically displayed in the plot area of the char

* https://matplotlib.org/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='Squre of number')
axes.plot(x,x**3,label='Qube of Number')

#axes.legend(loc=0)    # Legend use label of axis , positio of Legend are presnet in  above html link
axes.legend(loc=(0.1,0.1))  # based on position of X and Y we can set the legends 
```
